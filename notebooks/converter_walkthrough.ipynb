# Converter Walkthrough
$$
\tilde{r}[n]=\frac{r[n]-128}{128},\;\mathcal{H}\{\tilde{r}\}\rightarrow z[n],\;s[m]=\text{resample}_{f_s\rightarrow f'_s}(z[n]e^{-j2\pi\Delta f n/f_s}),\;b=\mathcal{Q}_{u8}(\operatorname{interleave}(\Re s, \Im s)).
$$

$$
\Lambda: \mathcal{F}\rightarrow\text{LineTable},\quad \mathcal{F}=\texttt{scripts/convert\_dat\_to\_bin.py}.
$$

$$
\Lambda: \text{script}\mapsto\{(\ell_i, s_{\ell_i})\}_{i=1}^{M},\quad \text{view lines via }\texttt{show\_lines}.
$$

```python
from pathlib import Path
SCRIPT = Path('../scripts/convert_dat_to_bin.py')
LINES = SCRIPT.read_text().splitlines()

def show_lines(start, end):
    for no in range(start, end+1):
        print(f"{no:04d}: {LINES[no-1]}")
```

## Fraction check
$$
\rho = \frac{f'_s}{f_s} = \frac{2048}{26000} = \frac{128}{1625}.\\
\text{Implementation uses }\operatorname{Fraction}(f'_s,f_s)\text{ to get }(\text{up},\text{down}).
$$

```python
from fractions import Fraction
import yaml
cfg = yaml.safe_load(Path('../configs/pipeline.yaml').read_text())
ratio = Fraction(int(cfg['signal']['fs_out']), int(cfg['signal']['fs_in']))
print('fs_in:', cfg['signal']['fs_in'], 'fs_out:', cfg['signal']['fs_out'])
print('up:', ratio.numerator, 'down:', ratio.denominator)
```
